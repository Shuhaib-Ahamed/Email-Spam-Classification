#### Overview of AI techniques used for email spam classification

Supervised learning algorithms such as Naive Bayes, Support Vector Machines (SVMs), Random Forest, and Gradient Boosting have been commonly used for email spam classification (RAZA, Jayasinghe and Muslam, 2021; Cota and Zinca, 2022; Singh et al., 2022). These algorithms are trained on labelled data and use the patterns in the labeled data to classify new emails as spam or non-spam. In the case of Naive Bayes, it is a probabilistic algorithm that makes assumptions about probability distributions based on past data. SVMs, on the other hand, are robust against overfitting due to their use of decision planes instead of straight lines, which provides good generalization performances even with small datasets (Cota and Zinca, 2022). Random Forest is an ensemble method that combines predictive models into larger decision trees instead of relying only on single instances based on previously trained models. Gradient Boosting uses multiple weak learners chosen randomly from different subsets.

In addition, unsupervised learning techniques such as clustering and anomaly detection have also been used for email spam classification (Awad, 2011). These methods do not rely on labeled data but instead identify data patterns to classify emails as spam or non-spam. Clustering algorithms such as K-means and DBSCAN can be used to group emails based on their characteristics (RAZA, Jayasinghe and Muslam, 2021), and anomaly detection algorithms such as one-class SVM and Random Forest can identify emails that differ from the norm.

### Naive Bayes 

Naive Bayes is a popular probabilistic technique in machine learning and natural language processing. It is useful for text classification and spam filtering since it can be easily trained on a big text dataset and can handle multiple features. The fundamental advantage of Naive Bayes is its simplicity and ease of implementation, which makes it computationally efficient. This makes it a popular choice for applications with low computational capabilities, such as mobile devices or embedded systems.

The ability of Naive Bayes to handle high-dimensional data, which is typical in text data, is its main strength. It can also handle missing data, which is a prevalent problem in text classification assignments. It may also estimate the probability of a specific class, which is useful in some situations.

However, some flaws in Naive Bayes should be considered for classification problems. One of the key flaws is the assumption of feature independence, which may not always be true in real-world data. This can result in poor performance and incorrect predictions. Furthermore, it can be sensitive to irrelevant features, and including unrelated data can result in poor performance. This is why feature selection is so important when using Naive Bayes.

Another drawback of Naive Bayes is that it is less flexible than other algorithms like Support Vector Machines (SVMs) or Neural Networks. These algorithms can learn more complex decision boundaries, improving specialized tasks' performance.

Naive Bayes requires labeled data for training and can handle both numerical and categorical variables in input data. The anticipated class label for a given input is the expected output.


### Support Vector Machine 

Support Vector Machines (SVMs) are supervised learning algorithms suitable for classification and regression problems. SVMs are designed to discover the best boundary, or hyperplane, that divides data into different classes. The boundary optimizes the distance between it and the closest data points in each category. This keeps the decision border as far away from the nearest points as possible, resulting in improved generalization and a lesser chance of overfitting.

When the data is not linearly separable, SVMs can be employed with a kernel trick to transfer the data into a higher-dimensional space where it becomes linearly separable. This capability enables SVMs to handle non-linearly separable and high-dimensional data, making it a valuable tool for a wide range of real-world applications.

Another benefit of SVMs is that they work well with limited datasets. This is because SVMs can discover the optimal boundary even when there are few data points, making it a good approach for situations with limited data. SVMs also contain a regularisation parameter that can be used to prevent overfitting, which is a prevalent problem in many machine-learning algorithms.

The biggest shortcoming of SVMs is that the algorithm is sensitive to the kernel and regularisation parameters. These factors substantially impact the model's performance, and determining the ideal values for these parameters can be difficult. Furthermore, SVMs can be computationally demanding when working with huge datasets, rendering them unsuitable for some applications.

Regarding input data, SVMs require labeled data for training and can handle both numerical and categorical variables; however, numerical data is preferred. In classification tasks, the expected output is a predicted class label for a given input, whereas, in regression tasks, the desired output is a projected value for a given information.

### Random Forest 

Random Forest is a well-known ensemble learning method that combines numerous decision trees to arrive at a final prediction. Random Forest works on the principle of creating numerous decision trees during training, each of which is trained on a different subset of the data. The random forest uses the majority vote of all decision trees to create a final forecast during prediction. This approach reduces overfitting by averaging the predictions of numerous decision trees and improves the model's overall accuracy.

Random Forest's key advantage is that it can handle high-dimensional data and categorical variables well. This is because decision trees can handle both numerical and categorical inputs, and Random Forests can successfully combine several decision tree forecasts to generate a final prediction. Random Forest can also be used for classification and regression tasks, making it a versatile method that can be utilised to solve a variety of issues.

Random Forest also has the advantage of being less prone to overfitting than a single decision tree. Because the many decision trees of a Random Forest are trained on independent data subsets, the correlation between the trees is reduced, and the risk of overfitting is reduced. Random Forest also offers a feature importance measure that may be used to determine which features are most relevant for making predictions, which might be useful in some applications.

However, Random Forest's biggest shortcoming is that it can be computationally expensive when working with large datasets. This is due to the fact that the algorithm must train and merge numerous decision trees, which can be computationally demanding. Furthermore, because Random Forest is a black box model that does not provide insight into the decision-making process, it is difficult to comprehend.

Random Forest requires labelled data for training and can handle both numerical and categorical variables in terms of input data. In classification tasks, the expected output is a predicted class label for a given input, whereas in regression tasks, the expected output is a projected value for a given input.

### Data

The [Enron-Spam datasets](https://www2.aueb.gr/users/ion/data/enron-spam/) (Metsis et. al., 2006) provided the data needed for the classification. All six preprocessed, malware-free datasets were collected. There are 785,648 instances total, each with a spam-detection indication. Each instance consists of an email sent by one of the six Enron employees. Examples include 407,140 (51.82%) tagged as ham and 378,508 (48.18%) as spam (non-spam). Each email's word count is disproportionately right-skewed, with an average of 60 words, The majority of the emails are 60 to 75 words long.

### Data Pre-Processing


*   To increase the accuracy of predictions, meaningless punctuation and stop words were removed from each email instance.
*   Randomly sampled 32,192 records to reduce computational burden.
*   Tokenized the original email contents into a list of unique words.
*   Vectorized the list of unique words with [TF-IDF](http://www.tfidf.com/) vectorizer.
*   Train-test split with a 0.2 test size ratio.

The following process combines various raw information into a single data frame. There are two columns in the data frame:

1.   Column 1: Label, Spam indicator (0 = not Spam; 1 = Spam)
2.   Column 2: Feature, Email 


```python
import pandas as pd
import csv
import glob
import re
import os
```

```python
name_of_raw_hams = [name for name in glob.glob(r'./data/raw/ham/*.txt')]
name_of_raw_spams = [name for name in glob.glob(r'./data/raw/spam/*.txt')]
list_of_csv_ham = []
list_of_csv_spam = []
```

```python
# Combine ham emails to list_of_csv_ham
for ham in name_of_raw_hams:
    ham_content = pd.read_csv(ham , engine='python' , sep='delimiter', header=None ,encoding="latin_1")
    list_of_csv_ham.append(ham_content)
    # Combine spam emails to list_of_csv_spam
for spam in name_of_raw_spams:
    spam_content = pd.read_csv(spam, engine='python', sep='delimiter', header=None ,encoding="latin_1")
    list_of_csv_spam.append(spam_content)
    # Convert to data frames
df_hams = pd.concat(list_of_csv_ham).reset_index(drop=True)
df_spams = pd.concat(list_of_csv_spam).reset_index(drop=True)
# Add the label column
df_hams['y'] = [0] * len(df_hams)
df_spams['y'] = [1] * len(df_spams)
```

```python
# Update column names
col_name = {0: 'X', 'y': 'y'}
df_hams = df_hams.rename(columns=col_name)
df_spams = df_spams.rename(columns=col_name)
```

```python
# Concatenate to a full dataset
df_emails = pd.concat([df_hams, df_spams], axis=0)
```

```python
# Save as csv files
df_hams.to_csv('./data/ham_emails.csv', index=False)
df_spams.to_csv('./data/spam_emails.csv', index=False)
df_emails.to_csv('./data/emails.csv', index=False)
```

#### Data Cleaning

```python
import pandas as pd
import string
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_csv("./data/emails.csv")
```

```python
# Randomly shuffle rows to mix ham emails with spam ones
df = df.sample(frac = 1).reset_index(drop=True)
```

```python
display(df.shape)
display(df.head(10))
display(df.tail(10))
```

```python
# Create a stop list
self_define = ['enron','subject','ect','hou','e','http'] # Manual list of words to be removed
stoplist = stopwords.words('english') + list(string.punctuation) + self_define
stoplist = set(stoplist)
```

```python
def trim_word(text):
    '''Remove unrelated words or symbols in emails
    Param: text: email content as a string
    '''
    text = [word for word in word_tokenize(text) if word.lower() not in stoplist and not word.isdigit() and word.isalpha()]
    return " ".join(text)
```

```python
df['X'] = df['X'].astype(str)
```

```python
# Remove punctuation, stop words, and self-defined words in X
df['X'] = df['X'].apply(trim_word)
```

```python
df2 = df.copy()
```

```python
# Check missing values
nan_value = float('NaN')
df2.replace('', nan_value, inplace=True)
n_missing_values = df2.isnull().sum()[0]
print('Total missing values (NaN) in the feature column:', n_missing_values)
print('\nTotal missing values (NaN) takes up {:.2%} of our data.'.format(n_missing_values/len(df.index)))
```

```python
# Remove rows containing missing values
df2.dropna(subset=['X'], inplace=True)
```

```python
df2 = df2.sort_values(by='X')
```

```python
# Drop rows that contain non-english characters
df2.drop(df2.tail(93).index, inplace=True)
```

```python
display(df2.shape)
display(df2.head(10))
display(df2.tail(10))
```

```python
# Save as cleand df
df2.to_csv('./data/emails_cleaned.csv', index=False)
```
